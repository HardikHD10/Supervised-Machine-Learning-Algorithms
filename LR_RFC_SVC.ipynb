```python
X = data.drop(['Diabetic'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest, Ytrain, Ytest = train_test_split(X,Y,train_size=0.7, test_size=0.3,stratify=Y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
model_1 = LogisticRegression()
model_2 = RandomForestClassifier()
model_3 = SVC()
```

```python
model_1.fit(Xtrain,Ytrain)
```

```python
model_2.fit(Xtrain,Ytrain)
```

```python
model_3.fit(Xtrain,Ytrain)
```

```python
pred_1 = model_1.predict(Xtest)
pred_2 = model_2.predict(Xtest)
pred_3 = model_3.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
```

```python
acry_1 = accuracy_score(Ytest,pred_1)
acry_2 = accuracy_score(Ytest,pred_2)
acry_3 = accuracy_score(Ytest,pred_3)
```

```python
print("Accuracy of LR : ",acry_1)
print("Accuracy of RandomForestClassifier : ",acry_2)
print("Accuracy of SVC : ",acry_3)
```

```python
print(classification_report(Ytest,pred_1))
```

```python
print(classification_report(Ytest,pred_2))
```

```python
print(classification_report(Ytest,pred_3))
```
